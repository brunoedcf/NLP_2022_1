```python
import nltk
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.corpus import wordnet as wn
nltk.download('reuters')
from nltk.corpus import reuters

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
cats = reuters.categories()
print("Reuters has %d categories:\n%s" % (len(cats), cats))

fileids = reuters.fileids()
```

```python
categories = []
text = []

for file in fileids:
    categories.append(reuters.categories(file))
    text.append(reuters.raw(file))

df = pd.DataFrame({'ids':fileids, 'categories':categories, 'text':text})

df
```

```python
text_list = df['text']

stopwords_nltk = stopwords.words('english')

from nltk.stem.porter import *
stemmer = PorterStemmer()

text_list_new = []
thisset = set()
emojiDict = {':)' : 'happy', '(:' : 'happy', ':(': 'sad', '):': 'sad', '(;' : 'wink', ';)' : 'wink', ':D' : 'happy', 'D:': 'sad'}
lemmatizer = nltk.stem.WordNetLemmatizer()
wn_lemmas = set(wn.all_lemma_names())

for t in text_list:
    # substituindo os emojis por palavras
    for word, initial in emojiDict.items():
      t = t.replace(word.lower(), initial)
    
    # regex para tirar caracteres, numeros, upcode e links htpp
    words = re.sub(r'[\s!%*~\^´`=+<\[\]?&$:;@#.0-9()\/\"\'_-]+|[\s]*[\\]u[a-z0-9]{4}[\s]*|[\s]*http[^\s]+[\s]*', " ", t.lower())
    words = re.sub(r'\brt\b', "", words)
    words = re.sub(r'[\\]+', " ", words)
    words = words.strip()
    
    phrase = ""
    words = words.split()

    # tirando stopwords
    # colocando apenas as palavras que estao na gramatica wordnet
    # fazendo stematizacao
    for x in words:
        if x not in stopwords_nltk and x in wn_lemmas:
            x = stemmer.stem(x)
            thisset.add(x)
            phrase = phrase + " " + x
        

    text_list_new.append(phrase.strip())
```

```python
text_list_new[0]
```

```python
vocab = set()

for s in text_list_new:
  sent = s.split()
  for w in sent:
    vocab.add(w)
  
vocab = sorted(vocab)
vocab
```

```python
bow_ponderado = {}
N = {} # quantidade de termos em cada documento

for i, doc in enumerate(text_list_new):
    bow_ponderado[i+1] = dict()
    N[i+1] = len(doc.split())
    for word in doc.split():
        if word in bow_ponderado[i+1]:
            bow_ponderado[i+1][word] += 1
        else:
            bow_ponderado[i+1][word] = 1

for doc in bow_ponderado:
    for word in bow_ponderado[doc]:
        bow_ponderado[doc][word] /= N[doc]
```

```python
bow_ponderado
```

```python
import math
idf = {}
for word in vocab:
    idf[word] = 0
    #contabilizar em quantos documentos a palavra word aparece
    for doc in bow_ponderado:
        if word in bow_ponderado[doc]:
            idf[word] += 1

for word in idf:
    idf[word] = math.log(len(text_list_new)/idf[word])

idf
```

```python
bow_tfidf = {}

for doc in bow_ponderado:
    bow_tfidf[doc] = dict()
    for word in bow_ponderado[doc]:
        bow_tfidf[doc][word] = bow_ponderado[doc][word]*idf[word]

bow_tfidf
```

```python
new_df = pd.DataFrame().from_records(bow_tfidf).fillna(0).T
new_df
```

```python
import math

def cosine_similarity(vetor1,vetor2):

    prod_interno = 0

    for i,valor in enumerate(vetor1):
        prod_interno += valor*vetor2[i]
    
    norma_vetor1 = 0
    for x in vetor1:
        norma_vetor1 += x*x
    norma_vetor1 = norma_vetor1**0.5

    #norma_vetor1 = math.sqrt(sum([x**2 for x in vetor1])) 

    norma_vetor2 = 0
    for x in vetor2:
        norma_vetor2 += x*x
    norma_vetor2 = norma_vetor2**0.5

    return prod_interno/(norma_vetor1*norma_vetor2)


```

```python
mat_cos = []


for i in range(1, 101):
  aux = []
  for j in range(1, 101):
      if i != j:
        aux.append( (cosine_similarity(new_df.iloc[i], new_df.iloc[j]), j) )

  aux.sort(reverse = True)
  mat_cos.append(aux[:10])

mat_cos

```

```python
top_ten = [0] * 10

for i, pos in enumerate(mat_cos):
  
  for j, s in enumerate(pos):
    for cat in df['categories'][i]:
      if cat in df['categories'][s[1]]:
        top_ten[j] += 1 
        

top_ten


```
